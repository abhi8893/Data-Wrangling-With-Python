# Exploring NYC 311 datasets  

```python
# because of mixed types we specify dtype to prevent any errors
complaints = pd.read_csv('data/311-service-requests.csv', dtype='unicode')
complaints.head()
```

## Cleaning the columns 

```python
from helper import ColumnNameCleaner
col_clnr = ColumnNameCleaner(lower=True)

clean_cols = col_clnr.clean(complaints)
complaints.columns = clean_cols
```

```python
complaints.head()
```

## Selecting only noise complaints 

```python
uniq_complaint_types = complaints['complaint_type'].unique()
```

```python
%%timeit

noise_cmplnt_types = []

for cmplnt in uniq_complaint_types:
    if 'noise' in cmplnt.lower():
        noise_cmplnt_types.append(cmplnt)
        
noise_cmplnt_types
```

### [Selecting elements in numpy array using regex](https://stackoverflow.com/questions/6595759/selecting-elements-in-numpy-array-using-regular-expressions)

```python
import re
```

```python
vmatch = np.vectorize(lambda x: bool(re.search('Noise', x, re.IGNORECASE)))
is_noise = vmatch(uniq_complaint_types)

noise_cmplnt_types = uniq_complaint_types[is_noise]
noise_cmplnt_types
```

```python
noise_cmplnt_types
```

```python
is_noise_complaint = np.isin(complaints['complaint_type'], noise_cmplnt_types)
is_noise_complaint
```

Using a for loop

```python
is_noise_complaint_2 = []

for comp in complaints['complaint_type']:
    if comp in noise_cmplnt_types:
        is_noise_complaint_2.append(True)
    else:
        is_noise_complaint_2.append(False)
        
is_noise_complaint_2 = np.array(is_noise_complaint_2)
```

```python
np.alltrue(is_noise_complaint == is_noise_complaint_2)
```

```python
complaints.loc[is_noise_complaint, :].head()
```

## Noise complaints in Brooklyn 

```python
in_brooklyn = complaints['borough'] == 'BROOKLYN'
noise_in_brooklyn = in_brooklyn & is_noise_complaint

complaints[noise_in_brooklyn].head()
```

Select only a few columns

```python
cols_to_select = ['complaint_type', 'borough', 'created_date', 'descriptor']

complaints[noise_in_brooklyn][cols_to_select]
```

## Which borough has the most noise? 

```python
noise_counts_borough = complaints.loc[is_noise_complaint, 'borough'].value_counts()
noise_counts_borough.name = 'borough_counts'
noise_counts_borough
```

```python
noise_counts_borough.plot(kind='bar')
```

```python
noise_prop_borough = noise_counts_borough/complaints['borough'].value_counts()
noise_prop_borough.name = 'borough_prop'
noise_prop_borough
```

```python
noise_prop_borough.plot(kind='bar')
```

```python
fig, axn = plt.subplots(2, 1, figsize=(12, 8))

noise_counts_borough.plot(kind='bar', ax=axn[0])
noise_prop_borough.plot(kind='bar', ax=axn[1])
```

```python
borough_noise_cmplts_mrgd = pd.concat([noise_counts_borough, noise_prop_borough], axis=1).reset_index()
borough_noise_cmplts_mrgd.rename(columns={'index': 'borough'}, inplace=True)
borough_noise_cmplts_mrgd
```

```python
fig, axn = plt.subplots(2, 1, figsize=(12, 8))

borough_noise_cmplts_mrgd['borough_counts'].plot(kind='bar', ax=axn[0])
borough_noise_cmplts_mrgd['borough_prop'].plot(kind='bar', ax=axn[1])

for ax, col in zip(axn, ['counts', 'prop']):
    colname = f'borough_{col}'
    borough_noise_cmplts_mrgd[colname].plot(kind='bar', ax=ax)
    ax.set_xticklabels(borough_noise_cmplts_mrgd['borough'], rotation=360)
    ax.set_title(col, fontdict=dict(weight='bold'))
    
fig.suptitle('Comparision of noise counts and complaints', fontdict=dict(weight='bold', size=40));
```
